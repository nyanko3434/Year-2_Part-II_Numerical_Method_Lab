```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def _visualize_bisection(func, a, b, midpoints):
    x = np.linspace(a, b, 400)
    y = func(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='f(x)', color='blue')
    plt.axhline(0, color='black', linewidth=0.5)

    for i, c in enumerate(midpoints):
        plt.axvline(c, color='red', linestyle='--', alpha=0.3)
        plt.plot(c, func(c), 'ko')  # Mark the midpoint

    plt.title("Bisection Method Visualization")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#functions for bisectional method in Numerical Methods
def bisection(func, a, b, tol=1e-5, max_iter=100, visualize=True):
    if func(a) * func(b) >= 0:
        raise ValueError("Function values at the endpoints must have opposite signs.")
    
    iterations = []
    midpoints = []

    for i in range(max_iter):
        c = (a + b) / 2.0
        iterations.append(i)
        midpoints.append(c)

        if abs(func(c)) < tol or (b - a) / 2 < tol:
            if visualize:
                _visualize_bisection(func, a, b, midpoints)
            return c

        if func(c) * func(a) < 0:
            b = c
        else:
            a = c

    if visualize:
        _visualize_bisection(func, a, b, midpoints)
    
    raise ValueError("Maximum iterations reached without convergence.")
```

```python
def bisection_with_steps(func, a, b, tol=1e-5, max_iter=100):

    if func(a) * func(b) >= 0:
        raise ValueError("Function values at the endpoints must have opposite signs.")
    
    print(f"Initial interval: [{a}, {b}]")
    
    for i in range(max_iter):
        c = (a + b) / 2.0  # Midpoint
        print(f"Iteration {i+1}: c = {c}, f(c) = {func(c)}")
        
        if abs(func(c)) < tol or (b - a) / 2 < tol:
            return c  # Root found
        
        if func(c) * func(a) < 0:
            b = c  # Root is in left half
            print(f"New interval: [{a}, {b}]")
        else:
            a = c  # Root is in right half
            print(f"New interval: [{a}, {b}]")
    
    raise ValueError("Maximum iterations reached without convergence.")
```

```python
def func_one(x):
    return x**2 - 4
```

```python
def func_two(x):
    return x**3 - x - 2
```

```python
def func_three(x):
    return x**3 - 3*x + 2
```

```python
# Example usage:
# if __name__ == "__main__":
function_variable = func_one

try:
    root = bisection(function_variable, 0, 5, max_iter=1000)
    print(f"\n\nRoot found: {root}\n\n")
except ValueError as e:
    print(e)
```

```python
function_variable = func_two
try:
    root = bisection_with_steps(function_variable, 1, 2, max_iter=1000)
    print(f"\n\nRoot found: {root}\n\n")
except ValueError as e:
    print(e)
```

```python
function_variable = func_three
try:
    root = bisection(function_variable, -50, 50, max_iter=10)
    print(f"\n\nRoot found: {root}\n\n")
except ValueError as e:
    print(e)
```
