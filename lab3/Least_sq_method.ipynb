```python
import math 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visulize_linear_LeastSquares(X, y, slope, intercept):

    plt.figure(figsize=(10, 6))


    plt.scatter(X, y, color='blue', label='Data points')


    y_pred = slope * X + intercept

 
    plt.plot(X, y_pred, color='red', label='Least Squares Fit')

    plt.xlabel('Feature')
    plt.ylabel('Target')
    plt.title('Least Squares Fitting')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def linear_Least_sq_method(X, Y, visualize=True):
    X = np.array(X)
    Y = np.array(Y)
    if len(X) != len(Y):
        raise ValueError("X and Y must have the same length")
    n = len(X)
    if n == 0:
        raise ValueError("X and Y cannot be empty")

    x_mean = np.mean(X)
    y_mean = np.mean(Y)
    numerator = np.sum((X - x_mean) * (Y - y_mean))
    denominator = np.sum((X - x_mean) ** 2)
    if denominator == 0:
        raise ValueError("Denominator is zero, cannot compute slope")
    slope = numerator / denominator
    intercept = y_mean - slope * x_mean
    if visualize:
        visulize_linear_LeastSquares(X, Y, slope, intercept)
    print(f"y = {slope:.4f} * x {' - ' if intercept > 0 else ' + '} {abs(intercept):.4f}")
    return slope, intercept
```

```python
X = [1, 2, 3, 4, 5]
Y = [2, 3, 5, 7, 11]
slope, intercept = linear_Least_sq_method(X, Y)

print(f"Slope: {slope}, Intercept: {intercept}")
```

```python
def visualize_exponential_LeastSquares(X, Y, slope, intercept):
    plt.figure(figsize=(10, 6))
    plt.scatter(X, Y, color='blue', label='Data points')
    y_pred = np.exp(intercept) * np.exp(slope * X)
    plt.plot(X, y_pred, color='red', label='Exponential Fit')
    plt.xlabel('Feature')
    plt.ylabel('Target')
    plt.title('Exponential Least Squares Fitting')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def exponential_Least_sq_method(X, Y, visualize=True):
    X = np.array(X)
    Y = np.array(Y)

    log_Y = np.log(Y)

    if len(X) != len(log_Y):
        raise ValueError("X and Y must have the same length")
    n = len(X)
    if n == 0:
        raise ValueError("X and Y cannot be empty")

    x_mean = np.mean(X)
    y_mean = np.mean(log_Y)
    numerator = np.sum((X - x_mean) * (log_Y - y_mean))
    denominator = np.sum((X - x_mean) ** 2)
    if denominator == 0:
        raise ValueError("Denominator is zero, cannot compute slope")
    a = numerator / denominator
    b = y_mean - a * x_mean
    if visualize:
        visualize_exponential_LeastSquares(X, Y, a, b)
    print(f"log(y) = {a:.4f} * x {' - ' if b > 0 else ' + '} {abs(b):.4f}")
    return a, b
```

```python
X = [1, 2, 3, 4, 5]
Y = [2.718, 7.389, 20.085, 54.598, 148.413]
slope, intercept = exponential_Least_sq_method(X, Y)
print(f"Exponential Fit - Slope: {slope}, Intercept: {intercept}")
```
