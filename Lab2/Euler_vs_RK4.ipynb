```python
import math 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize_Euler_vs_RK4(dfunc, x_euler, y_euler, x_rk4, y_rk4):
    plt.figure(figsize=(12, 6))
    plt.plot(x_euler, y_euler,label="Euler's Method", color='blue', linestyle='dotted')
    plt.plot(x_rk4, y_rk4,label="RK4 Method", color='red', linestyle='--')
    plt.title("Euler's Method vs RK4 Method")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
    for i in range(len(x_euler)):
        print(f"x: {x_euler[i]:.2f}, Euler y: {y_euler[i]:.4f}, RK4 y: {y_rk4[i]:.4f}, Difference: {abs(y_euler[i] - y_rk4[i]):.4f}")
```

```python
def RK4_Method(df, x0, y0, h = 0.1, max_itera = 200, visualize = True):
    x_values = [x0]
    y_values = [y0]

    x = x0
    y = y0

    for _ in range(max_itera):
        k1 = h * df(x, y)
        k2 = h * df(x + h / 2, y + k1 / 2)
        k3 = h * df(x + h / 2, y + k2 / 2)
        k4 = h * df(x + h, y + k3)

        y = y + (1/6) * (k1 + 2*k2 + 2*k3 + k4)
        x = x + h

        x_values.append(x)
        y_values.append(y)
    
    # if visualize:
    #     visualize_RK4_Method(df,x_values, y_values)

    return x_values, y_values
```

```python
def Euler_Method(dfunc, x, y, h=0.1, max_itera = 200, visualize = True):
    x_list = [x]
    y_list = [y]
    for _ in range(max_itera):
        y+= h * dfunc(x,y)
        x+= h
        x_list.append(x)
        y_list.append(y)
    
    # if visualize:
    #     visualize_Euler_Method(dfunc, x_list, y_list)
    return x_list, y_list
```

```python
def Euler_vs_RK4(dfunc, x0, y0, h=0.1, max_itera=50):
    x_euler, y_euler = Euler_Method(dfunc, x0, y0, h, max_itera)
    x_rk4, y_rk4 = RK4_Method(dfunc, x0, y0, h, max_itera)

    visualize_Euler_vs_RK4(dfunc, x_euler, y_euler, x_rk4, y_rk4)
```

```python
def f(x, y):
    return x * math.exp(-y)

x0 = 0
y0 = 1


h = 0.1
max_itera = 10

Euler_vs_RK4(f, x0, y0, h, max_itera)
```
