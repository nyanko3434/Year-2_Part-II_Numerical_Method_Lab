```python
import math 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize_Euler_Method(dfunc, x, y):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'o-', label='Euler Approximation', color='blue')

    # if exact_solution:
    #     x_exact = np.linspace(x0, x_end, 100)
    #     y_exact = [exact_solution(xi) for xi in x_exact]
    #     plt.plot(x_exact, y_exact, label='Exact Solution', color='green')

    plt.title("Euler's Method vs Exact Solution")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#Euler Method to find aproximate solution of a differential equation
def Euler_Method(dfunc, x, y, h=0.1, max_itera = 200, visualize = True):
    x_list = [x]
    y_list = [y]
    for _ in range(max_itera):
        y+= h * dfunc(x,y)
        x+= h
        x_list.append(x)
        y_list.append(y)
    
    if visualize:
        visualize_Euler_Method(dfunc, x_list, y_list)
    return x_list, y_list
```

```python
def dfunc_one(x, y):
    return x + y
```

```python
# Example usage
x = 0
y = 1
try:
    solution_x, solution_y = Euler_Method(dfunc_one, x ,y, max_itera=10)
    for xi, yi in zip(solution_x, solution_y):
        print(f"x = {xi:.1f}, y ≈ {yi:.5f}")
except ValueError as e:
    print(e)
```
